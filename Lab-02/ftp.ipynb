```python
from ftplib import FTP
import os

# FTP server config (dlptest.com provides free public FTP server)
CONFIG = {
    'host': 'ftp.dlptest.com',
    'port': 21,
    'username': 'dlpuser',
    'password': 'rNrKYTX9g7z3RgJRmxWuGHbeu',
}

LOCAL_UPLOAD = 'sample_upload.txt'
LOCAL_DOWNLOAD = 'downloaded_sample.txt'

# Step 1: Connect to FTP server
def ftp_connect(cfg):
    ftp = FTP()
    ftp.connect(cfg['host'], cfg['port'], timeout=20)
    ftp.login(cfg['username'], cfg['password'])
    ftp.set_pasv(True)
    print(f"‚úÖ Connected to FTP server {cfg['host']}")
    return ftp

# Step 2: Upload file
def upload_file(ftp, local_path, remote_name=None):
    if remote_name is None:
        remote_name = os.path.basename(local_path)
    with open(local_path, 'rb') as f:
        ftp.storbinary(f'STOR {remote_name}', f)
    print(f"‚úÖ Uploaded: {local_path} -> {remote_name}")

# Step 3: Download file and verify content
def download_file(ftp, remote_name, local_path):
    with open(local_path, 'wb') as f:
        ftp.retrbinary(f'RETR {remote_name}', f.write)
    print(f"‚úÖ Downloaded: {remote_name} -> {local_path}")

    # Verify content
    with open(local_path, 'r', encoding='utf-8') as f:
        content = f.read().strip()
    print("üìÑ Downloaded file content:")
    print(content)

# Step 4: List directory contents
def list_dir(ftp):
    print("üìÇ Directory contents on server:")
    files = ftp.nlst()
    for f in files:
        print("-", f)

if __name__ == "__main__":
    # Create a sample file for upload
    with open(LOCAL_UPLOAD, 'w', encoding='utf-8') as f:
        f.write("This is a sample file for FTP assignment test.\n")

    try:
        ftp = ftp_connect(CONFIG)

        list_dir(ftp)                                # Step 4
        upload_file(ftp, LOCAL_UPLOAD)               # Step 2
        download_file(ftp, os.path.basename(LOCAL_UPLOAD), LOCAL_DOWNLOAD)  # Step 3
        list_dir(ftp)                                # Step 4 again (to confirm upload)

        ftp.quit()
        print("‚úÖ FTP session closed.")
    except Exception as e:
        print("‚ùå FTP Error:", e)
```
